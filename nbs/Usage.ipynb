# Usage of the Storage Box AI API

#### Measure 2D Boxes (Also posts detections on mongo collection)

```python
import requests
import pandas as pd

# Define the URL and parameters
base_url = 'http://localhost:5000/process'

params = {
    # 'url': 1,
    'url': 'rtsp://admin:141291@octateste.ddns-intelbras.com.br:554/cam/realmonitor',
    'channel': 4,
    'subtype': 0,
    'max_frames': 301,  # (default: 300)
    'process_each': 30,  # (default: 90)
    # 'yield_each': 1,  # (default: 30)
    'conf': 0.00001, # (default: 0.001)
    'yield_type': 'results',
    # 'imgsz': 640, # (default: 640)
    # 'classes': '20',  # (default: '20') 'classes' should be a string with comma-separated values if there are multiple classes
    # 'max_det': None,  # (default: None)  # or you can leave this out if you don't want to set it,
    'fill_results': 'none', # 'frame-by-frame' (default) or 'none'
    # 'filter_results': 'processed-only', # 'processed-only' (default) or 'none'
}

# Make the GET request to the endpoint
response = requests.get(base_url, params=params)

# Check if the request was successful
if response.status_code == 200:
    # Print the JSON response
    results = response.json()
    df = pd.DataFrame(results)

    print("Results:")
    display(df)
else:
    print("Failed to get results:", response.status_code, response.text)
```
